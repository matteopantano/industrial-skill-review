```python
from sentence_transformers import SentenceTransformer

# initialize sentence transformer model
#model = SentenceTransformer('bert-base-nli-mean-tokens') #this is outdated: https://huggingface.co/sentence-transformers/bert-base-nli-mean-tokens
# info: https://www.sbert.net/docs/pretrained_models.html
model = SentenceTransformer('all-mpnet-base-v2') # SentenceTransformer('all-mpnet-base-v2')
sentence_embeddings = model.encode(defSingA)
print(sentence_embeddings.shape)
```

### Clustering

```python
n_neighbors = 10
n_components = 2

import umap
umap_embeddings = umap.UMAP(n_neighbors=n_neighbors, 
                            n_components=n_components, 
                            metric='cosine').fit_transform(sentence_embeddings)
```

Now clusters are created

```python
from sklearn.cluster import KMeans

clusterA = KMeans(n_clusters=10, n_init=200, random_state=0).fit(umap_embeddings)
```

```python
d = {'primitive': defSingA, 'cluster': resultA['labels'], 'x': resultA['x'], 'y': resultA['y']}

clusterDF_A = pd.DataFrame(d)
clusterDF_A['Doc_ID'] = range(0,len(clusterDF_A))

wordsPerTopicA = clusterDF_A.groupby(['cluster'], as_index = False).agg({'primitive': ' '.join})
```

As suggested by this [guide](https://towardsdatascience.com/keyword-extraction-with-bert-724efca412ea) we get the most recurring keywords for each cluster

```python
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial import distance
from nltk.corpus import wordnet as wn

# group by "X" column
groups = clusterDF_A.groupby(['cluster'])
  
labelsA = []

# extract keys from groups
keys = groups.groups.keys()

for i in keys:

    g = groups.get_group(i)
    n_gram_range = (1, 1)
    stop_words = "english"

    words = g['primitive'].to_list()

    count = CountVectorizer(ngram_range=n_gram_range, stop_words=stop_words).fit(words)
    candidates = count.get_feature_names()

    model = SentenceTransformer('all-mpnet-base-v2') # msmarco-distilbert-base-tas-b
    doc_embedding = model.encode(words)
    candidate_embeddings = model.encode(candidates)

    top_n = 2
    distances = cosine_similarity(doc_embedding, candidate_embeddings)
    keywords = [candidates[index] for index in distances.argsort()[0][-top_n:]]

    print("GROUP: " + str(i))
    
    print(keywords)

    matchedKeywords = str(keywords[0] + "-" + keywords[1])
    # print(matchedKeywords)
    
    labelsA.append(matchedKeywords)
```

Now add keywords to each data frame

```python
# By knowing the names extracted from the clusters we can then assign them to the clusterDF which can plot the scatter plot
clusterDF_A["keyword"] = ""

# We add the keyword to the cluster
for idx, row in clusterDF_A.iterrows(): 
    for i in range(0,len(labelsA)): # clusters statr from 0
        if clusterDF_A.iloc[idx]["cluster"] == (i):  
            clusterDF_A.at[idx, "keyword"] = labelsA[i]
```

Now we try to create the Latex table

```python
# group by "X" column
groups = clusterDF_A.groupby('cluster')
  
# extract keys from groups
keys = groups.groups.keys()

citations_column_names = ["cluster", "task", "skill", "primitive"]
citationsCluster = pd.DataFrame(columns = citations_column_names)

row = 0

taskTableFile = open(os.path.join(this_file_dir, "..", "data/out/" + outputName + "/" +  "taskEntries_NONIND.tex"), "w") 
taskSkillFile = open(os.path.join(this_file_dir, "..", "data/out/" + outputName + "/" +  "skillEntries_NONIND.tex"), "w") 
taskPrimitiveFile = open(os.path.join(this_file_dir, "..", "data/out/" + outputName + "/" +  "primitiveEntries_NONIND.tex"), "w") 
for i in keys:
    entries = []
    authorsP = []
    authorsT = []
    authorsS = []

    taskTableFile.write(f"""{groups.get_group(i).iloc[0]["keyword"]}&""")
    taskSkillFile.write(f"""{groups.get_group(i).iloc[0]["keyword"]}&""")
    taskPrimitiveFile.write(f"""{groups.get_group(i).iloc[0]["keyword"]}&""")

    for id in groups.get_group(i)['Doc_ID']:
        #print(id)
        if id in indexesAT:
            pos = indexesAT.index(id)
            #print("HEY I AM A TSK: " + str(id) + " " + str(clusterDF_A.iloc[id]["primitive"]))
            #print(dataframeT.iloc[indexesT[pos]]["author"]) # name of the author related to a certain entry
            for entry in dataframeT.iloc[indexesT[pos]]["bibtex"]:
                if entry not in authorsT and entry != "-":
                    #print(entry)
                    taskTableFile.write(f"""\cite{{{entry}}},""")
                    authorsT.append(entry)
        
        if id in indexesAS:
            pos = indexesAS.index(id)
            #print("HEY I AM A SKILL: " + str(id) + " " + str(clusterDF_A.iloc[id]["primitive"]))
            #print(dataframeS.iloc[indexesS[pos]]["author"])  # name of the author related to a certain entry
            for entry in dataframeS.iloc[indexesS[pos]]["bibtex"]:
                if entry not in authorsS and entry != "-":
                    #print(entry)
                    taskSkillFile.write(f"""\cite{{{entry}}},""")
                    authorsS.append(entry)
        
        if id in indexesAP:
            pos = indexesAP.index(id)
            #print("HEY I AM A PRIMITIVE: " + str(id) + " " + str(clusterDF_A.iloc[id]["primitive"]))
            #print(dataframeP.iloc[indexesP[pos]]["author"].to_list()) # name of the author related to a certain entry
            for entry in dataframeP.iloc[indexesP[pos]]["bibtex"]:
                if entry not in authorsP and entry != "-":
                    #print(entry)
                    taskPrimitiveFile.write(f"""\cite{{{entry}}},""")
                    authorsP.append(entry)  
        
    citationsCluster.at[row, "cluster"] = groups.get_group(i).iloc[0]["keyword"]
    citationsCluster.at[row, "task"] = ",".join(str(x) for x in authorsT) 
    citationsCluster.at[row, "skill"] = ",".join(str(x) for x in authorsS) 
    citationsCluster.at[row, "primitive"] = ",".join(str(x) for x in authorsP) 

    taskTableFile.write(f"""\n""")
    taskSkillFile.write(f"""\n""")
    taskPrimitiveFile.write(f"""\n""")
    row += 1

#Save the author bibtek entry in somehow and array and then use the array for the automatic table creation

taskTableFile.close()
taskSkillFile.close()
taskPrimitiveFile.close()
```

We now create the latex table autoomatically using the information from citationsCluster

```python
# group by "X" column
groups = citationsCluster.groupby('cluster')
  
# extract keys from groups
keys = groups.groups.keys()

# table file
tableFile = open(os.path.join(this_file_dir, "..", "data/out/" + outputName + "/" +  "citationEntries_NONIND.tex"), "w") 

for i in keys:

    tableFile.write(f"""{groups.get_group(i).iloc[0]["cluster"]}&""")

    # task
    entries = groups.get_group(i)['task'].tolist()
    entries = entries[0].split(",")
    tableFile.write(f"""{{\\tiny""")
    for entry in entries:
        # last element in the list
        if entry == entries[-1]:
            if len(entry) > 0:
                tableFile.write(f"""\cite{{{entry}}} }}&""")
            else:
                tableFile.write(f"""&""")
        else:
            tableFile.write(f"""\cite{{{entry}}},""")

    # skill
    tableFile.write(f"""{{\\tiny""")
    entries = groups.get_group(i)['skill'].tolist()
    entries = entries[0].split(",")
    for entry in entries:
        # last element in the list
        if entry == entries[-1]:
            tableFile.write(f"""\cite{{{entry}}} }}&""")
        else:
            tableFile.write(f"""\cite{{{entry}}},""")

    # primitive
    tableFile.write(f"""{{\\tiny""")
    entries = groups.get_group(i)['primitive'].tolist()
    entries = entries[0].split(",")
    for entry in entries:
        # last element in the list
        if entry == entries[-1]:
            tableFile.write(f"""\cite{{{entry}}} }}""" r"""\\""")
        else:
            tableFile.write(f"""\cite{{{entry}}},""")
    
    tableFile.write(f"""\n""")
    tableFile.write(f"""\hline\n""")

tableFile.close()
    
```
